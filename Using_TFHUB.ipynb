Import libraries

```
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import PIL.Image as Image
```

Get the url for the classifier from the tfhub

```
classification_url = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/classification/2" #@param {type:"string"}
```

Fix image shape

```
IMAGE_SHAPE = (224,224)
```

Convert tfhub model as keras model(wrapping)

```
classifier = tf.keras.Sequential([
      hub.KerasLayer(classification_url,input_shape = IMAGE_SHAPE + (3,))
])
```

Get any custom image

```
grace_hopper = tf.keras.utils.get_file('image.jpg','https://storage.googleapis.com/download.tensorflow.org/example_images/grace_hopper.jpg')
grace_hopper = Image.open(grace_hopper).resize(IMAGE_SHAPE)
plt.imshow(grace_hopper)
grace_hopper = np.array(grace_hopper) / 255.0
```

predicting image

```
result = classifier.predict(grace_hopper[np.newaxis, ...])
predicted_class = np.argmax(result[0],axis = -1)
print(predicted_class)
```
